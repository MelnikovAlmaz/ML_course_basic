## Что такое Data Science и зачем он нужен?

**Data Science** (наука о данных) - это раздел информатики, изучающий проблемы анализа, обработки и визуализации данных.

## Основные понятия

Машинному обучению находят применение во многих сферах. А что же такое машинное обучение? Чтобы продолжить дальнейшее изучение темы, необходимо остановиться и разобраться в базовых вещах.

**Artificial intelligence (AI)** - искусственный интеллект (ИИ). Наука и технология создания интеллектуальных машин, особенно интеллектуальных алгоритмов и компьютерных программ.

**Machine learning (ML)** - машинное обучение (МЛ). Раздел искусственного интелекта (ИИ), объединяющий методы построения алгоритмов, которые адаптируются для решения конкретной задачи на основе исходного набора данных.

**Data analysis** - анализ данных. Это процесс проверки, очистки, преобразования и моделирования данных с целью обнаружения полезной информации, обоснования выводов и поддержки принятия решений.

## Процесс разработки
Перед тем как появится готовое приложение, например предсказывающее счет футбольного матча, должны быть пройдены определенные этапы ее разработки.

Даже когда привозят домой детали мебели, в коробку кладут инструкцию по сборке. Если Использовать инструкцию, то мебель будет собрана правильно и быстро, иначе вместо шкафа может получиться кровать, будет потрачено время и настроение будет не веселое. Также и в Data Science есть стандарты(инструкции) по созданию проектов.

В 2000 году был представлен **CRISP_DM** - межотраслевой стандартный процесс для исследования данных.

<img align="center" src="crispdm.jpg" width="500"></img>

Согласно CRISP-DM, проект проект состоит из шести основных этапов, выполняемых последовательно:
 * Бизнес-анализ (Business understanding)
 * Анализ данных (Data understanding)
 * Подготовка данных (Data preparation)
 * Моделирование (Modeling)
 * Оценка результата (Evaluation)
 * Внедрение (Deployment)

## Профессии в Data Science

Создание проекта состоит из нескольких этап и на каждом из них подключаются разные специалисты. У каждой из них свои особенности и инструменты, которые они применяют.

В основном выделяют 4 профессий:
 * **Специалист по организации данных** - продумывает какие данные будут хранится в базе, следит за тем чтобы данные были четкими и никуда не терялись. Использует базы данных Hadoop, PostgreSQL, Oracle.
 * **Дата инженер** - он является связующим звеном между данными и всеми остальными специалистами, которым данные нужны. Он поднимает сервера для получения и обработки данных из базы, такие как Hadoop, Spark, Hive.
 * **Бизнес аналитик** - человек-презентация. Он анализирует данные, находит в них интересную информацию и демонтрирует результаты людям, принимающим важные решения, в основном это директора. Использует дэшборды для получения данных, например Tableau, также использует сервисы, которые настроил Дата инженер.
 * **Специалист по машинному обучению** - тот самый человек, который исследует данные, создает и обучает алгоритмы, работает с нейронными сетями. Использует в работе Python, библиотеки машинного обучения Scikit-learn, для нейронных сетей Keras и Tensorflow.
