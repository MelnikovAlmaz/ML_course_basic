## Описание данных

Таблица содержит 506 строк и 14 столбцов.

Данные содержат следующие столбцы:
 - **crim** - уровень преступности на душу населения по городам.
 - **zn** - доля жилой земли, зональной для участков более 25 000 кв. футов.
 - **indus** - доля непродаваемых акров в городе.
 - **chas** -1, если улица пересекает реку, 0 в противном случае).
 - **nox** - концентрация оксидов азота.
 - **rm** - среднее количество комнат на одно жилье.
 - **age** - процент занятых домов от количества домов, построенных до 1940.
 - **dis** - взвешенное среднее расстояний до пяти центров занятости в Бостоне.
 - **rad** - показатель доступности до кольцевых магистралей.
 - **tax** - полная стоимость налога на имущество на каждые 10000 дол.
 - **ptratio** - соотношение учеников и учителей по городу.
 - **black** - 1000(Bk - 0.63)^2 где Bk - доля чернокожих по городу.
 - **lstat** - процент населения низкого статуса.
 
## Задача

Предсказать цену дома по информации о доме.

```python
# Импортируем необходимые библиотеки
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
   
# Подгружаем данные
from sklearn.datasets import load_boston 
boston = load_boston()
```

```python
boston.data.shape
```

```python
boston.feature_names
```

```python
# Смотрим на первые 10 записей
data = pd.DataFrame(boston.data) 
data.columns = boston.feature_names 
  
data.head(10)
```

```python
# Adding 'Price' (target) column to the data  
boston.target.shape
```

```python
data['Price'] = boston.target 
data.head() 
```

```python
# Смотрим на базовые статистики по данным
data.describe()
```

```python
# Подготавливаем данные для обучения модели
x = boston.data 
   
# Подготавливаем целевую переменную 
y = boston.target 
   
# Разделяем данные на обучающую и тренировочную выборки  
from sklearn.model_selection import train_test_split 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2, 
                                                    random_state = 0) 
   
print("xtrain shape : ", xtrain.shape) 
print("xtest shape  : ", xtest.shape) 
print("ytrain shape : ", ytrain.shape) 
print("ytest shape  : ", ytest.shape) 
```

```python
# Обучаем линейную регрессию 
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression() 
regressor.fit(xtrain, ytrain) 
   
# Предсказываем значения цены на тестовой выборке
y_pred = regressor.predict(xtest) 
```

```python
# Строим график предсказанной цены к реальной цене
plt.scatter(ytest, y_pred, c = 'green') 
plt.xlabel("Цена: в $1000") 
plt.ylabel("Предсказанное значение") 
plt.title("Реальное значение к предсказанному : Линейная регрессия") 
plt.show() 
```
