## Введение в анализ данных и машинное обучение

Данные называют нефтью 21 века. Открытие нефти изменило большинство сфер нашей жизни. Из нее производят автомобили, медицинское оборудование, а также Айфоны. Также и данные изменят нашу жизнь. Чтобы извлекать пользу из данных их также нужно уметь добывать и обрабатывать.

### Применение

<img align=center src="amazon.jpg" style="height:100px;" />

Интернет магазины, такие как Amazon или Aliexpress, ежедневно собирают огромное количество информации о покупках пользователей и действиях, которые предшествовали им. Зная, что заказывает клиент, система может порекомендовать и другие товары, которые могут ему понравится. Это также приносит пользу бизнесу, увеличивая кол-во проданных товаров.

<img align=center src="rentgen.jpg" style="height:100px;" />

А вот еще задача. Медицинские учреждения, которые  делают снимки легких для обнаружения болезней, хранят тысячи изображений. Алгоритмы машинного обучения уже сейчас с большей точностью чем врачи  могут распознать болезнь на ранней стадии. Благодаря этому жизнь человека может быть спасена, вовремя назначив ему лечение.

<img align=center src="ingineer.jpg" style="height:100px;" />

Машинное обучение также применяется в авиации. Ежедневно через аэропорты пролетают сотни самолетов. Самолет - это сложная техника, в которой установлены тысячи датчиков. Человек просто не в состоянии просмотреть все эти показатели и понять, где может произойти поломка. А вот алгоритм может быстро обработать большой объем данных, выявить нестабильную работу деталей и предупредить о возможной поломке. Вовремя устраненная неисправность может предотвратить авиакрушение.

### Основные понятия

Машинному обучению находят применение во многих сферах. А что же такое машинное обучение? Чтобы продолжить дальнейшее изучение темы, необходимо остановиться и разобраться в базовых вещах.

**Data Science** - наука о данных. Это раздел информатики, изучающий проблемы анализа, обработки и визуализации данных.

**Artificial intelligence (AI)** - искусственный интеллект (ИИ). Наука и технология создания интеллектуальных машин, особенно интеллектуальных алгоритмов и компьютерных программ.

**Machine learning (ML)** - машинное обучение (МЛ). Раздел искусственного интелекта (ИИ), объединяющий методы построения алгоритмов, которые адаптируются для решения конкретной задачи на основе исходного набора данных.

**Data analysis** - анализ данных. Это процесс проверки, очистки, преобразования и моделирования данных с целью обнаружения полезной информации, обоснования выводов и поддержки принятия решений.
